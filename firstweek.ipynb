```python
# 1. Car Class Example
# This class represents a car with brand and model information.

class Car:
    def __init__(self, brand, model):  # Constructor method to initialize brand and model
        self.brand = brand
        self.model = model

    def display_info(self):  # Method to print car information
        print(f"Brand: {self.brand}, Model: {self.model}")

# Usage example
car1 = Car("Toyota", "Corolla")  # Creating a Car object
car1.display_info()  # Printing the car's information
```

```python
# 2. Rectangle Class Example
# This class represents a rectangle and calculates its area.

class Rectangle:
    def __init__(self, width, height):  # Constructor to initialize width and height
        self.width = width
        self.height = height

    def calculate_area(self):  # Method to calculate and return area
        return self.width * self.height

# Usage example
rectangle1 = Rectangle(5, 10)  # Creating a Rectangle object
area = rectangle1.calculate_area()  # Calculating area
print(f"Rectangle area: {area}")
```

```python
# 3. Square Class Example
# This class represents a square and prints the square using '*'.

class Square:
    def __init__(self, side_length):  # Constructor to initialize side length
        self.side_length = side_length

    def print_square(self):  # Method to print the square, no return value
        for i in range(self.side_length):
            print('* ' * self.side_length)

# Usage example
square1 = Square(5)  # Creating a Square object
square1.print_square()  # Printing the square
```

```python
# 4. Calculator Class Example
# This class can sum two or three numbers depending on the input.

class Calculator:
    def add(self, num1, num2, num3=None):  # Method to add two or three numbers
        if num3 is not None:
            return num1 + num2 + num3
        else:
            return num1 + num2

# Usage example
calculator = Calculator()

# Adding two numbers
result1 = calculator.add(10, 20)
print("Sum of two numbers:", result1)

# Adding three numbers
result2 = calculator.add(10, 20, 30)
print("Sum of three numbers:", result2)
```

```python
# 5. Hello Class Example
# This class prints 'Hello World' without requiring any user input.

class Hello:
    def say_hello(self):  # Method to print Hello World
        print("Hello World")

# Usage example
hello = Hello()  # Creating a Hello object
hello.say_hello()  # Calling the method
```

```python
# 6. People and Derived Classes Example
# `Person` base class and derived classes: `Teacher`, `Secretary`, and `Student`.

# Base class: Person
class Person:
    def __init__(self, name, age):  # Constructor with name and age
        self.name = name
        self.age = age

    def speak(self):  # General speaking method
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Derived class: Teacher
class Teacher(Person):
    def __init__(self, name, age, id_number):
        super().__init__(name, age)
        self.id_number = id_number  # Unique ID for teacher

    def speak(self):  # Overridden speak method
        print(f"Hello, I am teacher {self.name}, age {self.age}.")

    def give_lesson(self):
        print(f"Teacher {self.name} is giving a lesson.")

# Derived class: Secretary
class Secretary(Person):
    def speak(self):  # Overridden speak method
        print(f"Hello, I am secretary {self.name}, ready to help.")

    def organize_meeting(self):
        print(f"Secretary {self.name} is organizing a meeting.")

# Derived class: Student
class Student(Person):
    def speak(self):  # Overridden speak method
        print(f"Hi, I am student {self.name}, age {self.age}.")

    def attend_class(self):
        print(f"Student {self.name} is attending the class.")

# Usage examples
teacher1 = Teacher("Alice", 40, 101)
teacher1.speak()
teacher1.give_lesson()

secretary1 = Secretary("Bob", 35)
secretary1.speak()
secretary1.organize_meeting()

student1 = Student("Charlie", 20)
student1.speak()
student1.attend_class()
```
